The **compression ratio** is defined as the ratio between the volume of the cylinder with the piston in the bottom position, Vbottom (largest volume), and in the top position, Vtop (smallest volume). The higher this ratio, the greater will be the power output from a given engine. It is generally in the **610 range**. 

We have few outlier that are in range of 20's , however after getting some background knowlege it was understood that disesel engines range is between **18-23**


**RPM** stands for revolutions per minute, and its used as a measure of how fast any machine is operating at a given time. In cars, rpm measures how many times the engines crankshaft makes one full rotation every minute, and along with it, how many times each piston goes up and down in its cylinder.
A typical engine rev-range is from an idle of between **500 and 1000rpm**, and a maximum speed of about 6000 to 7000rpm.



**City-mpg and highway-mpg** seems reasonable because typically, it is in the range of 20's-30's. However highway-mpg is usally is higher that city-mpg.

Engine-Type: Categorical/Nominal

For engine-type 

      I can do LabelEncoder Transformation since this is a categorical data 








```python

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler





df1=df
scaler= StandardScaler()
scaler= scaler.fit(df1['symboling'].values.reshape(-1,1))
std_feature=scaler.transform(df1['symboling'].values.reshape(-1,1))
dataframe1=pd.DataFrame(std_feature)

plt.hist(dataframe1)
plt.title('Tranforming of symboling of dataset')
plt.show()
# print('symboling has been transformed to fit the model' ,xx1)
print("\n")

df2=df



le=preprocessing.LabelEncoder()
le.fit(df2['engine-type'].values.reshape(-1,1))
array=le.transform(df2['engine-type'].values.reshape(-1,1))
# print(df2.head())
xx1=pd.unique(array)
print('Uniquely encoded each engine-type brand to a number ',xx1)

print("\n")


df3=df
le=preprocessing.LabelEncoder()
le.fit(df3['make'].values.reshape(-1,1))
array2=le.transform(df3['make'].values.reshape(-1,1))
dataframe2=pd.DataFrame(array2)
xx=pd.unique(array2)
print('Uniquely encoded each car brand to a number ',xx)






```
